```python
files = glob.glob("/content/drive/My Drive/DBJ2021/*.txt")
dat_list = []
BS_list = []
for file in files:
    if '貸借対照表' not in file:
        print(file)
        dat = pd.read_table(file, sep = ',')
        dat_list.append(dat)
    else:
        print(file)
        dat = pd.read_table(file, sep = ',')
        BS_list.append(dat)
same_col = []
if len(BS_list) == 2:
    col1 = BS_list[0].columns.tolist()
    col2 = BS_list[1].columns.tolist()
    for i in col1:
        if i in col2:
            same_col.append(i)
    dat = pd.merge(BS_list[0], BS_list[1], on = same_col, how = 'outer')
    dat_list.append(dat)
    print("指定されたフォルダ内のファイルの読み込みは完了しました。")
else:
    print("BSの結合を別途やり直してください。")
```

```python
BS_col = dat_list[2].columns.tolist()
PL_col = dat_list[1].columns.tolist()
CashFlow_col = dat_list[0].columns.tolist()
def delete_space(x):
    x = x.replace("　", " ").replace(" ", "")
    return(x)
for k in BS_col:
    k.replace("　", " ").replace(" ", "")
for k in PL_col:
    k.replace("　", " ").replace(" ", "")
for k in CashFlow_col:
    k.replace("　", " ").replace(" ", "")
capital_col = []
debt_col = []
profit_col = []
list1 = [name for name in BS_col if '資本' in name]
for col_name in list1:
    col_set = [col_name, 2, BS_col.index(col_name)]
    capital_col.append(col_set)
list2 = [name for name in BS_col if '負債' in name]
for col_name in list2:
    col_set = [col_name, 2, BS_col.index(col_name)]
    debt_col.append(col_set)
list2_sub = [name for name in BS_col if 'RTF0007000' in name]
list2_sub_sub = 'RTF0007000  負債合計'
debt_col_sub_set = [list2_sub_sub, 2, BS_col.index(list2_sub[0])]
debt_col.append(debt_col_sub_set)
list3 = [name for name in PL_col if '損益' in name]
for col_name in list3:
    col_set = [col_name, 1, PL_col.index(col_name)]
    profit_col.append(col_set)
ROE_col = capital_col + debt_col + profit_col
for i in range(len(ROE_col)):
    ROE_each_col = ROE_col[i]
    i += 1
    ROE_each_col = [i] + ROE_each_col
    ROE_col[i-1] = ROE_each_col
for j in range(len(ROE_col)):
    print(ROE_col[j])
```

```python
dif_list = []
for i in range(len(dat_list[1])):
    if (dat_list[1].iat[i,1] != dat_list[2].iat[i,1]) | (dat_list[1].iat[i,3] != dat_list[2].iat[i,3]):
        l = [i, "違うよ"]
        dif_list.append(l)
if len(dif_list) == 0:
    print("BSとPLのrowは完全一致")
else:
    print("BSとPLのrowの順序が違うから注意")
```

```python
necessary_col = [23,24,38, 39]
#上の出力から必要な要素の先頭番号(1~46)を[]内に　,　で区切って入れていく
ser1 = dat_list[1]['会社名']
ser2 = dat_list[1]['決算期']
ROE_dat = pd.DataFrame({'会社名': ser1, '決算期': ser2})
li = []
for i in range(len(dat_list[1])):
    i += 1
    li.append(i)
for num in necessary_col:
    for j in range(len(ROE_col)):
        if num == ROE_col[j][0]:
            dat_num = ROE_col[j][2]
            column_num = ROE_col[j][3]
            column_name = ROE_col[j][1]
            new_list = dat_list[int(dat_num)].iloc[:, int(column_num)].tolist()
            new_ser = pd.Series(new_list, index = li, name = column_name)
            new_dat = pd.DataFrame(new_ser)
            ROE_dat = pd.concat([ROE_dat, new_dat], axis=1)
```

```python
ROE_dat['RSK0010700  （旧）当期純損益'] = ROE_dat['RSK0010700  （旧）当期純損益'].tolist()
ROE_dat['RSK0010750  当期純損益'] = ROE_dat['RSK0010750  当期純損益'].tolist()
ROE_dat['RTF0007000  負債合計'] = ROE_dat['RTF0007000  負債合計'].tolist()
ROE_dat['RTF0011000  負債及び純資産合計'] = ROE_dat['RTF0011000  負債及び純資産合計'].tolist()
```

```python
Profit_new_True = ROE_dat['RSK0010700  （旧）当期純損益'].isnull().tolist()
Profit_old = ROE_dat['RSK0010700  （旧）当期純損益'].tolist()
Profit_new = ROE_dat['RSK0010750  当期純損益'].tolist()
profit = []
for i in range(len(Profit_new)):
    if Profit_new_True[i] == True:
        profit.append(Profit_new[i])
    else:
        profit.append(Profit_old[i])
ROE_dat['当期純損益'] = profit
```

```python
all_debt = ROE_dat['RTF0011000  負債及び純資産合計'].tolist()
debt = ROE_dat['RTF0007000  負債合計'].tolist()
profit_null = ROE_dat['当期純損益'].isnull().tolist()
roe = []
for i in range(len(profit)):
    if profit_null[i] == True:
        roe.append(np.nan)
    else:
        roe.append(100 * profit[i]/(all_debt[i]-(debt[i])))
ROE_dat['ROE(%)'] = roe
ROE_dat.sample(50)
```
